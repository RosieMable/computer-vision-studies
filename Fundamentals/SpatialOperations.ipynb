<h2> Spatial Operations in Image Processing</h2>

Spatial operations use  pixels in a neighborhood to determine the present pixel value. Applications include filtering and sharpening. They are used in many steps in computer vision like segmentation and are a key building block in Artificial Intelligence algorithms.

- Linear Filtering
    - Filtering Noise
    - Gaussian Blur
    - Image Sharpening
- Edges
- Median
- Threshold

# Convolution (Linear Filtering)

Convolution is a mathematical operation used to filter images. It involves a kernel (or filter) that is applied to each pixel and its neighbors.

## Process
1. The kernel is overlaid on a region of the image.
2. Each element of the kernel is multiplied by the corresponding pixel value.
3. The results are summed to produce a new pixel value in the output image.
4. This process is repeated by shifting the kernel across the entire image.

# Edge Detection

Edge detection is crucial for identifying boundaries within images, which is a fundamental step in many computer vision algorithms.
It approximates derivatives and gradients to find areas where brightness changes sharply.

The Sobel operator  is commonly used to compute horizontal and vertical gradients.
The result is an image that highlights the edges, represented as vectors indicating the direction of intensity changes.

# Median Filters

Median Filters are effective in removing certain types of noise whiler preserving edges better than mean filters.
For each pixel, the median value of the surrounding neighborhood is calculated and used as the new pixel value; this helps in reducing noise without signficantly distoring the image.

# Low Pass Filters

These filters smooth the image by averaging pixel values, which helps in reducing noise.
Trade-off: While they effectively reduce noise, they can also make edges less sharp, leading to a loss of detail.
