# Image Classification with KNN 

## K-Nearest Neighbor (KNN)
- KNN is a simple classification algorithm that classifies unknown data points by finding the most common classes among the k-nearest examples.
- It identifies the closest match based on distance.

## Example of Classification
- Imagine two classes: dogs and cats, represented on a 2D plane.
- Each image is labeled (y = 0 for cat, y = 1 for dog).
- The images are converted into vectors for processing.

## Distance Calculation
- The **Euclidean distance** is used to measure the distance between two images.




- A table is created with training samples, their classes, and distances to an unknown sample.


## Prediction Process
- The label of the unknown sample is predicted based on the nearest neighbor.
- The predicted label is denoted as $\hat{Y}$ (the hat indicates an estimate).

## Model Evaluation
- Data is split into **training** and **testing** sets to evaluate model performance.
- Typically, 70% of the data is used for training and 30% for testing.
- Accuracy is calculated by comparing predicted values with actual class labels.

## Majority Voting
- If the nearest sample is ambiguous (e.g., a cat next to a dog), KNN uses the **K closest samples** and performs a majority vote to assign the label.

## Choosing K
- The optimal value of K is determined using a **validation set**.
- The dataset is split into training, validation, and testing sets.
- The K that maximizes accuracy on the validation set is selected.

## Real-World Application
- Once K is chosen, KNN can classify images (e.g., recognizing cats and dogs).
- The app calculates distances, finds nearest neighbors, and outputs the class.

## Limitations of KNN
- KNN can be slow and may struggle with challenges in image classification.
- The video suggests exploring other classifiers for more efficient solutions.
